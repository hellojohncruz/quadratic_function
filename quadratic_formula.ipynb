```python
import matplotlib.pyplot as plt
```

```python
'''
Quadratic function calculator and graph
'''

quad_pts = list()

# Assume values of x
x_values = range(-10,10)
for x in x_values:
    # Calculate the value of the quadratic function
    y = x**2 + 2*x + 1
    quad_pts.append((x,y))
    print(f'x={x}, y={y}')
plt.plot(*zip(*quad_pts))
plt.show()
```
